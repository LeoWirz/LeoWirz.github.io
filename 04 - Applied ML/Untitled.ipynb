```python
import pickle
def save_obj(obj, name ):
    with open('obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open('obj/' + name + '.pkl', 'rb') as f:
        return pickle.load(f)
    
test = load_obj('gridsearch_result')
```

```python
import numpy.ma as ma
from matplotlib import cm

depths = ma.getdata(test['param_max_depth']).reshape(4,-1)
trees = ma.getdata(test['param_n_estimators']).reshape(4,-1)
scores = test['mean_test_score'].reshape(4,-1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(depths, trees, scores, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

ax.set_xlabel('depth')
ax.set_ylabel('estimators')
ax.set_zlabel('mean score')

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random

def fun(x, y):
  return x**2 + y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
y = np.arange(-3.0, 3.0, 1)
x = np.arange(-2,2,1)
X, Y = np.meshgrid(x, y)
zs = np.array([fun(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
